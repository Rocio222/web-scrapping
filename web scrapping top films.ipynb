### Top films Filmaffinity

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://www.filmaffinity.com/es/topcat.php?id=new_th_es"

response = requests.get(url)
print(bool(response))
soup = BeautifulSoup(response.text, "html.parser")
```

```python
tit=soup.find_all('h3')
titulos=[i.text.strip() for i in tit]
```

```python
dur=soup.find_all("div",class_="duration")
duracion=[i.text.strip() for i in dur]
```

```python
pun=soup.find_all("div",class_="avg-rating")
puntuacion=[i.text.strip() for i in pun]
```

```python
gen=soup.find_all("a",class_="genre")
genero=[i.text.strip() for i in gen]
```

```python
dire=soup.find_all("span",class_="nb")
director=[i.text.strip() for i in dire]
```

```python
cas=soup.find_all("span",class_="nb")
casting=[i.text.strip() for i in cas]
```

```python
vot=soup.find_all("span",class_="nb")
casting=[i.text.strip() for i in cas]
```

```python
vot=soup.find_all("div",class_="rat-count")
votaciones=[i.text.strip() for i in vot]
```

```python
df=pd.DataFrame(columns=["Titulo","Duracion","Puntuacion","Genero","Votacion"])
df["Titulo"]=titulos
df["Duracion"]=duracion
df["Puntuacion"]=puntuacion
df["Genero"]=genero
df["Votacion"]=votaciones
```
